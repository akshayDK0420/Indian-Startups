# Data Cleaning.

```python
df.head()
```

Droping column unnamed and description.

```python
df.drop(columns=['Unnamed: 0','Description'],axis = 1,inplace = True)
```

```python
df['Industries Type 1'] = df['Industries'].str.split(',',expand=True)[0]
```

```python
df['Industries Type 2'] = df['Industries'].str.split(',',expand=True)[1]
```

Creating a column to check how old the company is.

```python
df['Age'] = df['Starting Year'].max()-df['Starting Year']
```

```python
sns.distplot(df['Age'])
```

Dividing the age of the company into 3 segments.

```python
df['Age_Category'] = pd.cut(df.Age,[-1,5,15,np.inf],labels = ['New Company','Old Company','Older Company'])
```

```python
df.sample()
```

# Exploratory data analysis and Insights.

In which year maximum startups are launched.

```python
plt.figure(figsize=(15,5),dpi=100)
sns.countplot(x = 'Starting Year',data=df)
plt.xticks(rotation = 45)
plt.show()
```

Maximum startups were born in 2015.

In which city maximum startups are launched.

```python
from wordcloud import WordCloud as word
```

```python
wc = word(background_color='white')
wc.generate_from_frequencies(df['City'].value_counts())
plt.imshow(wc)
plt.axis('off')
plt.show()
```

Which founders were involved in 2 or more startups?

```python
x = df['Founders'].value_counts().sort_values(ascending=False).head(4)
x
```

```python
color = sns.color_palette('twilight_shifted')
plt.pie(x,labels=x.index,autopct='%0.0f%%',shadow = True,colors=color,explode=[0,0.1,0.1,0.1])
plt.show()
```

What were the No. of Employees in the startups.

```python
no_employess = df['No. of Employees'].value_counts().reset_index()
no_employess.columns = ('No. of Employees','Count')
no_employess
```

```python
x = no_employess['No. of Employees']
y = no_employess['Count']
plt.figure(figsize=(15,5),dpi=100)
plt.bar(x,y,color='grey')
plt.xticks(rotation = 45)
plt.title('No. of Employees')
for i in range(len(x)):
    plt.text(x[i],y[i],f'{y[i]}',ha='center',va='bottom',fontsize=8)
plt.grid()
plt.show()
```

Mostly Startups has between 101-250 of employess.

Which was the maximum funding round.

```python
funding_round = df['Funding Round'].value_counts().reset_index()
funding_round.columns = ('Funding Round','Count')
funding_round.sort_values(by='Funding Round',inplace=True)
funding_round
```

```python
x = funding_round['Funding Round']
y = funding_round['Count']
plt.figure(figsize=(15,5),dpi=100)
plt.bar(x,y,color='grey')
plt.title('Funding Rounds')
plt.grid()
for i in range(len(x)):
    plt.text(x[i],y[i],f'{y[i]}',ha='center',va='bottom',fontsize=8)
plt.show()
```
